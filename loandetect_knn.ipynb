```python
dftrain.info()
```

```python
x=['Gender','Married','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']
```

```python
train_X=dftrain[x]
train_y=dftrain['Loan_Status']
```

```python
train_y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(train_X)    # find out min & max value
```

```python
train_X=scaler.transform(train_X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
for i in range(3,12,2):
    model=KNeighborsClassifier(n_neighbors=i,weights="distance")
    model.fit(train_X,train_y)
    per=model.score(train_X,train_y)
    print(" neighbour is {} model accurancy is {}  ".format(i,per))
```
